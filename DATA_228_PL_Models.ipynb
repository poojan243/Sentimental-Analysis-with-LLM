```python
from textblob import TextBlob

def get_sentiment(text):
    analysis = TextBlob(text)
    polarity = analysis.sentiment.polarity

    if polarity > 0:
        return "Positive"
    elif polarity < 0:
        return "Negative"
    else:
        return "Neutral"
```

```python
# Input from the user
text_input = input("Enter a text string: ")
sentiment = get_sentiment(text_input)
print("Sentiment:", sentiment)
```

```python
data = pd.read_csv("/content/drive/MyDrive/Sem2/BigData/GA1-PL-MODELS/train.csv", encoding="ISO-8859-1")  # Replace with your actual dataset file

data['selected_text'] = data['selected_text'].astype(str)

def get_sentiment(text):
    analysis = TextBlob(text)
    polarity = analysis.sentiment.polarity

    if polarity > 0:
        return "positive"
    elif polarity < 0:
        return "negative"
    else:
        return "neutral"

# Apply sentiment analysis to each row in the dataset
data['analyzed_sentiment'] = data['selected_text'].apply(get_sentiment)

data.head(15)

# Save the updated dataset with sentiment analysis results
#data.to_csv("sentiment_analyzed_dataset.csv", index=False)  # Replace with desired output filename
```

```python
total_values = len(data['analyzed_sentiment'])

# Count the number of occurrences where "analyzed_sentiment" is different from "sentiment"
different_count = (data['analyzed_sentiment'] != data['sentiment']).sum()

# Count the number of occurrences where "analyzed_sentiment" is the same as "sentiment"
same_count = (data['analyzed_sentiment'] == data['sentiment']).sum()

percentage_difference = (different_count/total_values) * 100

print("Total values:", total_values)
print("Number of differences:", different_count)
print("Number of not different:", same_count)

print('Percentage of differences:', str(round(percentage_difference, 2)) + '%')
```
